````python
# 第二題ans
# 有一篇被加密過的文章片斷，
# 其加密的方法是將文章中的每個字母往後移"數"個字母後，組合而成加密後的文章片斷。
# 其中若有空格或其它符號，則保持不變，只根據英文字母進行加密
# 提示語: 未加密的文章片斷的最後一句: trouble will come of it
# 利用以上規則及提示語解密


# 先取得加密的檔案，並將它讀出來。
# 讀出來後，從password=1開始解密並取得解密後的字串s1
# 提示語s="trouble will come of it" 是文章句尾的最後一句(有空格，沒有其它符號)
# 從解密後的字串s1的最未端開始，取出字母及空格(排除其它所有符號)，取出的長度與s的長度相等
# 將上一動取出後的字串命名為s2
# 將s2倒轉過來
# 如果倒轉過來的s2=提示語s的話，那麼就找到密碼，而s1即是解密後的字串。

import urllib.request as req

url = "https://github.com/Elwing-Chou/tibame1224/raw/refs/heads/main/password.txt"
req.urlretrieve(url, "password.txt")
f = open("password.txt", "r", encoding="utf-8")
message = f.read()
f.close()

s = "trouble will come of it"

def find_s2(s1):
    s2 = ""
    i = 0
    while True:
        c = s1[-1 - i]
        if ord("a") <= ord(c) <= ord("z"):
            s2 = s2 + c
        elif c == " ":
            s2 = s2 + c
        if len(s2) == len(s):
            return s2
            break
        i = i + 1

def find_s1(message, password):
    s1 = ""
    for c in message.lower():
        if ord("a") <= ord(c) <= ord("z"):
            n = ord(c) - ord("a")
            n = (n - password) % 26
            s1 = s1 + chr(n + ord("a"))
        else:
            s1 = s1 + c
    return s1

password = 0
while True:
    password = password + 1
    s1 = find_s1(message, password)
    s2 = find_s2(s1)
    s2_reversed = s2[::-1]
    if s2_reversed == s:
        print("password= ", password)
        break


print("-" * 10 + "加密的文章片斷message" + "-" * 10)
print(message)
print("-" * 10 + "提示語s" + "-" * 10)
print(s)
print("-" * 10 + "解密後的文章片斷s1" + "-" * 10)
print(s1)
print("-" * 10 + "從s1的末端開始，取得與提示語s長度相等的字串s2" + "-" * 10)
print(s2)
print("-" * 10 + "將s2反轉無s2_reversed並確認是否與提示語s一致" + "-" * 10)
print(s2_reversed)


````
