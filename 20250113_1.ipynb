````python


import urllib.request as req
import bs4 as bs
# Step 0.
import pandas as pd

# Step1. 準備一個空的List
table = []
page = 1
while True:
    print("頁數:", page)
    url = "https://tabelog.com/tw/tokyo/rstLst/{}/?SrtT=rt".format(page)
    page = page + 1
    response = req.urlopen(url)
    html = bs.BeautifulSoup(response)

    r_list = html.find_all("div", {"class":"list-rst__body"})
    if len(r_list) == 0:
        break

    for r in r_list:
        en_name = r.find("a", {"class":"list-rst__rst-name-target"})
        en_name_href = en_name["href"]
        en_name_text = en_name.text
        area = r.find("div", {"class":"list-rst__area-genre"})
        area_text = area.text.strip()
        intro = r.find("p", {"class":"list-rst__pr-title"})
        if intro == None:
            intro_text = "無"
        else:
            intro_text = intro.text.strip()
        rating = r.find("span", {"class":"c-rating__val"})
        rating_text = rating.text
        prices = r.find_all("span", {"class":"c-rating-v3__val"})
        dinner_price = prices[0].text
        lunch_price = prices[1].text
        holiday = r.find("span", {"class":"list-rst__holiday-text"})
        holiday_text = holiday.text

        # print(rating_text)
        # print(en_name_text)
        # print(en_name_href)
        # print(area_text)
        # print(intro_text)
        # print(holiday_text)
        # print("晚間價錢:", dinner_price, "午間價錢:", lunch_price)
        # print("-" * 30)

        # Step2. 每一筆資料就是一個字典
        data = {
            "name(en)": en_name_text,
            "url": en_name_href,
            "area": area_text,
            "intro": intro_text,
            "rating": rating_text,
            "holiday": holiday_text
        }
        table.append(data)



# Step3. 轉成DataFrame儲存
# 字典是無序的，所以columns可能會亂掉，所以在轉成DataFrame時最好指定columns
df = pd.DataFrame(table,
          columns=["name(en)", "url", "area", "intro", "rating", "holiday"])
df.to_csv("tabelog.csv", encoding="utf-8")
print(df)
````

````python
practice = pd.DataFrame([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
], columns=["a", "b", "c"], index=["c", "d", "e"])
print(practice)
````

````python
# 取列(column, 欄，直的): 字典
print(practice["a"]) # 只取一列時，印出會顯示Name: a, dtype: int64，多列的不會顯示
print(practice[["a", "c"]]) # dict[[keys]]-> 把多個key包成list當成字典的查詢key
````

````python
# 只取一列時，印出會顯示Name: a, dtype: int64，多列的不會顯示
# pandas的2大資料型態:
# 1. Series(序列)型態: 類似list，單列，有行索引, 1D(1維)
# 2. DataFrame型態: 類似table, 有行有列，有行索引列索引 2D(2維)
# 補充，N-Dimensional 多維空間
# 三維空間是多個平面(2維)空間無限堆疊，四維空間是由多個立體(三維)空間無限堆疊
# 人類是活在三維空間裡，難以想像三維以上的空間


print(type(practice["a"])) # <class 'pandas.core.series.Series'>
print(type(practice[["a", "c"]])) # <class 'pandas.core.frame.DataFrame'>

print(practice["a"]["c"]) # 取出a這個序列，再用序列的行索引去取出單一值
print(practice["c"]["e"]) # 取出c這個序列，再用序列的行索引去取出單一值
print(type(practice["a"]["c"])) #<class 'numpy.int64'>


````

````python
# 取行(row, 橫的)
# 1. 看他實際第幾行(iloc): list,把0,1,2當key
# 2. 看他實際第幾行(loc): 字典,把行標籤當key


print(practice)
print("-" * 30)
print(practice.loc["d"])
print(practice.loc["e"])
print(practice.loc["c"]["c"]) # 用行索引c取出，再用列索引的欄位c取出單一值
print(practice.iloc[1])
print(practice.iloc[2])
print(practice.iloc[0]["c"])
print(practice.iloc[0:2])
print(practice.iloc[0:2]["c"])
````

````python

# 過濾(filter)原理: 把跟妳列的筆數一樣多的True False帶給她[]
# 對到True的就會留下, 對到False的就會丟掉
print(practice)
fil = [True, False, True]
print(practice[fil])
````
